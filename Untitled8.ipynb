```python
# =========================
# IMPORT LIBRARY
# =========================
import pandas as pd
import numpy as np
import re
import string

from sklearn.feature_extraction.text import CountVectorizer

# =========================
# MOUNT GOOGLE DRIVE
# =========================
from google.colab import drive
drive.mount('/content/drive')

# =========================
# LOAD DATASET (XLSX)
# =========================
path = "/content/drive/MyDrive/dataset_komentar.xlsx"

df = pd.read_excel(path)
print(df.head())

# =========================
# PREPROCESSING
# =========================

# Case Folding (huruf kecil semua)
df['clean'] = df['text'].str.lower()

# Hapus simbol & angka
def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', str(text))
    return text

df['clean'] = df['clean'].apply(clean_text)

# =========================
# STOPWORD REMOVAL
# =========================

!pip install Sastrawi

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()

def remove_stopwords(text):
    return ' '.join([word for word in text.split() if word not in stopwords])

df['clean'] = df['clean'].apply(remove_stopwords)

# =========================
# BAG OF WORDS (BoW)
# =========================

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['clean'])

# Fitur kata
print(vectorizer.get_feature_names_out())

# Bentuk matrix
print(X.shape)

# =========================
# UBAH KE DATAFRAME
# =========================

bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print(bow_df.head())

# Menampilkan semua kolom
pd.set_option('display.max_columns', None)
print(bow_df.head())

# =========================
# ANALISIS FREKUENSI KATA
# =========================

word_counts = np.sum(X.toarray(), axis=0)

freq_df = pd.DataFrame({
    'kata': vectorizer.get_feature_names_out(),
    'frekuensi': word_counts
}).sort_values(by='frekuensi', ascending=False)

print(freq_df.head(10))

# =========================
# VISUALISASI BAR CHART
# =========================

import matplotlib.pyplot as plt

top_n = 10
top_words = freq_df.head(top_n)

plt.figure(figsize=(10,5))
plt.bar(top_words['kata'], top_words['frekuensi'])
plt.title("Top 10 Kata Paling Sering Muncul")
plt.xticks(rotation=45)
plt.xlabel("Kata")
plt.ylabel("Frekuensi")
plt.show()

# =========================
# WORDCLOUD
# =========================

from wordcloud import WordCloud

# Gabungkan semua teks
all_text = ' '.join(df['clean'])

wordcloud = WordCloud(width=800, height=400).generate(all_text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("WordCloud Kata yang Sering Muncul")

plt.show()
```
